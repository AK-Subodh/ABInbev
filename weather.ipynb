## Feature Engineering

### Type Conversion

```python
## Before converting the date, we need to check the format of the date

data['date'].sample(10)
```

```python
## From above, we know that the format of the date is %m/%d/%Y

data['Date_parsed'] = pd.to_datetime(data['date'], format = "%m/%d/%Y")
data.head()
```

```python
data.head(10)
```

```python
data['time']=data['time']/100
data['time']=data['time'].astype('int')
data.head(10)
```

```python
data['Time_parsed'] = pd.to_datetime(data['time'], format = "%H")
data.head()
```

```python
data['month']=data['Date_parsed'].dt.month
data['year']=data['Date_parsed'].dt.year
data['day']=data['Date_parsed'].dt.day
data['hour']=data['Time_parsed'].dt.hour
data.head()
```

```python
data=data.drop(['Time_parsed','time','date'],axis=1)
data.head()
```

```python
data['season']=data['month']
data['season']=data['season'].replace([3,4,5],'spring')
data['season']=data['season'].replace([6,7,8],'summer')
data['season']=data['season'].replace([9,10,11],'autumn')
data['season']=data['season'].replace([12,1,2],'winter')
data.head()
```

```python
data.columns
```

As there are multiple columns assigned for the same attribute with just the different unit, including all of them in study makes no sense as they represent the same data. Also keeping them will just increase multicollinearity which is harmful for regression problems. 

Following is the list of columns with the same info:
- windspeedMiles & windspeedKmph
- precipMM & precipInches
- visibility & visibilityMiles
- pressure & pressureInches
- WindGustMiles & WindGustKmph

Let's drop following and work in metric system:
- windspeedMiles
- precipInches
- visibilityMiles
- pressureInches
- WindGustMiles

```python
data=data.drop(['windspeedMiles','precipInches','visibilityMiles','pressureInches','WindGustMiles'],axis=1)
data.head()
```
